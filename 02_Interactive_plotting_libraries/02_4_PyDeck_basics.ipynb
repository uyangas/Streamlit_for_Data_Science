```python
import pydeck as pdk
import streamlit as st
```

```python
data = [
    {'latitude': 47.9117868, 'longitude': 106.9117342, 'name': 'Lavazza'},
    {'latitude': 47.9126404, 'longitude': 106.8952637, 'name': 'Tom n Toms'},
    {'latitude': 47.9145791, 'longitude': 106.9058683, 'name': 'R.O.C'}
]
```

```python
layer = pdk.Layer(
    'ScatterplotLayer',
    data=data,
    get_position='[longitude, latitude]',
    get_radius=50,
    get_fill_color=[255, 0, 0], 
    pickable=True
)

# Газрын зургийг харуулах анхны байршил
view_state = pdk.ViewState(
    latitude=47.9146589,
    longitude=106.9073843,
    zoom=13
)

# Газрын зургыг render хийх
r = pdk.Deck(
    layers=[layer],
    initial_view_state=view_state
)
r.to_html('stores.html')
```

```python
import pandas as pd

data = pd.DataFrame(
    {'latitude': [47.918112,47.9144707,47.914784,47.913108,47.918258,47.9191905], 
     'longitude': [106.9197614,106.8900929,106.906200,106.8846104,106.932473,106.9414596],
     'name': ['Lavazza','Saraana','R.O.C','Ti Amo',"Jack's Coffee","Cafe Papa"],
     'sales':[100.0, 43.0, 135.0, 89.0, 67.0, 38.0]
    }
)
data
```

```python
layer = pdk.Layer(
    'GridCellLayer',
    data=data,
    get_position='[longitude, latitude]',
    get_elevation=['sales'],
    elevation_scale=5,
    extruded=True,
    get_radius=10,
    get_fill_color=['sales*10','sales','sales*40',220], 
    pickable=True,
    coverage=0.1
)

view_state = pdk.ViewState(
    latitude=47.9146589,
    longitude=106.9073843,
    pitch=50,
    zoom=13
)

# Газрын зургыг render хийх
r = pdk.Deck(
    layers=[layer],
    initial_view_state=view_state
)
r.to_html('stores.html')
```
